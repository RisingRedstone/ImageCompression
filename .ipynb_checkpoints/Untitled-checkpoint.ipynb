```python
%matplotlib inline  
import cv2
import numpy as np
import math
import matplotlib.pyplot as plt
from functions import RoundNum, HuffmanTree, QuantizationTablePreprocess
```

```python
img = cv2.imread("Images/Im1.png")
```

```python
img = img[:RoundNum(img.shape[0],8),:RoundNum(img.shape[1],8),:]
img = cv2.cvtColor(img, cv2.COLOR_BGR2YCR_CB)
DCTimg = np.zeros(img.shape)
```

```python
QTY = np.array([[16, 11, 10, 16, 24, 40, 51, 61],  # luminance quantization table
                [12, 12, 14, 19, 26, 48, 60, 55],
                [14, 13, 16, 24, 40, 57, 69, 56],
                [14, 17, 22, 29, 51, 87, 80, 62],
                [18, 22, 37, 56, 68, 109, 103, 77],
                [24, 35, 55, 64, 81, 104, 113, 92],
                [49, 64, 78, 87, 103, 121, 120, 101],
                [72, 92, 95, 98, 112, 100, 103, 99]])

QTC = np.array([[17, 18, 24, 47, 99, 99, 99, 99],  # chrominance quantization table
                [18, 21, 26, 66, 99, 99, 99, 99],
                [24, 26, 56, 99, 99, 99, 99, 99],
                [47, 66, 99, 99, 99, 99, 99, 99],
                [99, 99, 99, 99, 99, 99, 99, 99],
                [99, 99, 99, 99, 99, 99, 99, 99],
                [99, 99, 99, 99, 99, 99, 99, 99],
                [99, 99, 99, 99, 99, 99, 99, 99]])

QHighT = np.array([[20, 30, 40, 99, 99, 99, 99, 99],  # chrominance quantization table
                [30, 50, 60, 99, 99, 99, 99, 99],
                [40, 60, 99, 99, 99, 99, 99, 99],
                [99, 99, 99, 99, 99, 99, 99, 99],
                [99, 99, 99, 99, 99, 99, 99, 99],
                [99, 99, 99, 99, 99, 99, 99, 99],
                [99, 99, 99, 99, 99, 99, 99, 99],
                [99, 99, 99, 99, 99, 99, 99, 99]])
```

```python
QY = QuantizationTablePreprocess(QTY)
QC = QuantizationTablePreprocess(QTC)
```

```python
for x in [(8 * i) for i in range(int(img.shape[0]/8))]:
    for y in [(8 * i) for i in range(int(img.shape[1]/8))]:
        for i in range(8):
            for j in range(8):
                DCTimg[x+i,y+j,0] = int(np.sum(img[x:(x+8),y:(y+8),0] * QY.quant_pre[0][i, j]))
                DCTimg[x+i,y+j,1] = int(np.sum(img[x:(x+8),y:(y+8),1] * QC.quant_pre[0][i, j]))
                DCTimg[x+i,y+j,2] = int(np.sum(img[x:(x+8),y:(y+8),2] * QC.quant_pre[0][i, j]))
```

```python
InvDCTimg = np.zeros(img.shape)
```

```python
for x in [(8 * i) for i in range(int(img.shape[0]/8))]:
    for y in [(8 * i) for i in range(int(img.shape[1]/8))]:
        for i in range(8):
            for j in range(8):
                InvDCTimg[x+i,y+j,0] = np.sum(DCTimg[x:(x+8),y:(y+8),0] * QY.quant_pre[1][i, j])
                InvDCTimg[x+i,y+j,1] = np.sum(DCTimg[x:(x+8),y:(y+8),1] * QC.quant_pre[1][i, j])
                InvDCTimg[x+i,y+j,2] = np.sum(DCTimg[x:(x+8),y:(y+8),2] * QC.quant_pre[1][i, j])
```

```python
InvDCTimg[:,:,0] -= np.min(InvDCTimg[:,:,0])
#InvDCTimg[:,:,1] -= np.min(InvDCTimg[:,:,1])
#InvDCTimg[:,:,2] -= np.min(InvDCTimg[:,:,2])
```

```python
InvDCTimg = InvDCTimg.astype(np.uint8)
InvDCTimg = np.maximum(0, InvDCTimg)
InvDCTimg = np.minimum(255, InvDCTimg)
```

```python
fig, axis = plt.subplots(1, 2)

axis[0].imshow(cv2.cvtColor(img, cv2.COLOR_YCR_CB2RGB))
axis[0].set_title("Original Image")

axis[1].imshow(cv2.cvtColor(InvDCTimg, cv2.COLOR_YCR_CB2RGB))
axis[1].set_title("Regenerated Image")

plt.show()
```

```python
hist = [{}, {}, {}]
DCTimg = DCTimg.astype(np.uint8)
for x in range(DCTimg.shape[0]):
    for y in range(DCTimg.shape[1]):
        for c in range(DCTimg.shape[2]):
            hist[c][DCTimg[x,y,c]] = hist[c].get(DCTimg[x,y,c], 0) + 1
    
```

```python
H = [HuffmanTree() for _ in range(3)] 
H[0].insertDict(hist[0])
H[1].insertDict(hist[1])
H[2].insertDict(hist[2])
```

```python
encodings = [H[0].dict(), H[1].dict(), H[2].dict()]
EncodedStrings = [b'', b'', b'']

for x in range(DCTimg.shape[0]):
    for y in range(DCTimg.shape[1]):
        for c in range(DCTimg.shape[2]):
            EncodedStrings[c] += encodings[c][DCTimg[x,y,c]]
```
